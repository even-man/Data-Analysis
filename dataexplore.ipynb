```python
#Import libraries being used 
import pandas as pd
import string as str
```

```python
#read in csv file as a pandas dataframe
data_uncleaned = pd.read_csv('Data1.csv')
```

```python
#view columns in our dataframe and view first three rows to make sure our data is correct
print(data_uncleaned.columns)
data_uncleaned.head(3)
```

```python
#Drop columns that will not be needed for analysis
data_uncleaned = data_uncleaned.drop(columns=['Flight #','cn/In', 'Registration'], axis=1)
```

```python
#Create a copy of our dataframe called data_uncleaned_copy and create a new dataframe called New_Route_Data

data_uncleaned_copy = data_uncleaned

New_Route_Data = data_uncleaned_copy['Route'].str.split('-', expand=True)

#rename columns in New_Route_Data dataframe
New_Route_Data = New_Route_Data.rename(columns={0:'route1', 1:'route2', 2:'route3',3:'route4',4:'route5',5:'route6'})
New_Route_Data.head(3)
```

```python
#Break out location data into individual cells
New_Location_Data = data_uncleaned_copy['Location'].str.split(',', expand=True)
New_Location_Data = New_Location_Data.rename(columns={0:'location1', 1:'location2', 2:'location3',3:'location4'})
New_Location_Data[1000:1005]
```

```python
data_without_routes_or_location = data_uncleaned.drop(columns=['Route', 'Location'], axis=1)
data_without_routes_or_location.columns
```

```python
#Create new variables
data_without_routes_or_location["Total Deaths"] =data_without_routes_or_location['Fatalities'] + data_without_routes_or_location['Ground']
data_without_routes_or_location["Death Rate"] = data_without_routes_or_location['Fatalities'] / data_without_routes_or_location['Aboard']
data_without_routes_or_location["Survival Rate"] = 1 - data_without_routes_or_location['Death Rate']
data_without_routes_or_location[1000:1005]
```

```python
#Write dataframes to csv file 
data_without_routes_or_location.to_csv('data_without_routes_or_location.csv')
New_Route_Data.to_csv('New_Route_Data.csv')
New_Location_Data.to_csv('New_Location_Data.csv')
```
